```python

def fibonacci_steps_rounded_5000(n):
    fib_sequence = [1_000, 2_000]

    while fib_sequence[-1] < n:
        new_number = fib_sequence[-1] + fib_sequence[-2]
        rounded = np.ceil(new_number / 5000) * 5000
        fib_sequence.append(int(rounded))

    return fib_sequence[1:] 

short_run = np.arange(start=1_000, step=1_000, stop=20_000) 
fine_steps_25000 = np.arange(start=25_000, step=25_000, stop=1_000_000) 
fibonacci_ru_5000 = [2_000, 5_000, 10_000, 15_000, 25_000, 40_000, 65_000, 105_000, 170_000, 275_000, 445_000, 720_000, 1_165_000, 1_885_000]  #fibonacci_steps_rounded_5000(2_000_000)

interation_list = short_run
benchmark_dict_pandas = {}
benchmark_dict_polars = {}
start_time = time.time()

benchmark_df = pd.DataFrame(columns=['library', 'test', 'iterations', 'seconds'])

# TODO: Put other tests here.

def create_series_polars():
    enum_dtype = pl.Enum(["Polar", "Panda", "Brown"])
    enum_series = pl.Series(["Polar", "Panda", "Brown", "Brown", "Polar"], dtype=enum_dtype)

def create_series_pandas():
    enum_dtype = pd.CategoricalDtype(categories=["Polar", "Panda", "Brown"], ordered=True)
    enum_series = pd.Series(["Polar", "Panda", "Brown", "Brown", "Polar"], dtype=enum_dtype)

# pandas vs swifter vs polars
def run_tests(library, test_name, test, iterations):
    global start_time
    global benchmark_df
    print(f"New test {test.__name__} with {iterations} iterations.")
    
    for i in range(iterations):
        test()
    seconds = np.round(time.time() - start_time, 2)
    benchmark_df.loc[len(benchmark_df)] = [library, test_name, iterations, seconds]
    start_time = time.time()

test_dict = {
    "create_series": {
        "pandas": create_series_pandas,
        "polars": create_series_polars
    }
}

library_list = ['pandas', 'polars']
for test_name in test_dict:
    for iterations in interation_list:
        for library in library_list:
            test = test_dict[test_name][library]
            run_tests(library, test_name, test, iterations)
```

### Plot Results

```python
benchmark_df = pd.read_csv("benchmark_output.csv")
test_dict = {
    "create_series": {}
}

x_name = 'iterations'
y_name = 'seconds'

charts = []
for test_name in test_dict:  # One chart for each test
    chart = alt.Chart(benchmark_df).mark_point().encode(
        x=x_name,
        y=y_name,
        color='library:N',  # Color by library
    ).facet(
        column='test:N',  # Facet by test
        title=test_name
    )
    charts.append(chart)

combined_chart = alt.vconcat(*charts)  # Use vconcat for vertical concatenation
combined_chart.display()
```
